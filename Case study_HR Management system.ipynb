```python
class Employee:
    def __init__(self, department, name, birthday, address, onboard):
        self.department = department
        self.name = name
        self.birthday = birthday
        self.address = address
        self.onboard= onboard
        
from tkinter import *
from tkinter.messagebox import showinfo

#tạo màn hình
window = Tk()
window.title("HR management")
window.geometry("680x680")
header = "HR management"

employee_list = []

# Tạo label
title_lbl= Label(window, text=header, font=('Arial', 30), bd=3)
title_lbl.grid(row=0, column=0, columnspan=8, padx=20, pady=20)

department_lbl = Label(window, text="Department", font=('Arial', 12))
department_lbl.grid(row=1, column=0, padx=10,pady=10)
name_lbl = Label(window, text="Name", font=('Arial', 12))
name_lbl.grid(row=2, column=0, padx=10,pady=10)
birthday_lbl = Label(window, text="Birthday", font=('Arial', 12))
birthday_lbl.grid(row=3, column=0, padx=10,pady=10)
address_lbl = Label(window, text="Address", font=('Arial', 12))
address_lbl.grid(row=4, column=0, padx=10,pady=10)
onboard_lbl=  Label(window, text="Onboard", font=('Arial', 12))
onboard_lbl.grid(row=5, column=0, padx=10,pady=10)

#add employee
def add_employee():
    department_value = department.get()
    name_value = name.get()
    birthday_value = birthday.get()
    address_value = address.get()
    onboard_value = onboard.get()

    employee = Employee(department_value, name_value, birthday_value, address_value, onboard_value)
    employee_list.append(employee)

    showinfo("Result", "added successfully")

#search employee
def search_employee():
    search_value = search.get().strip()
    found = False  
    for employee in employee_list:
        if search_value.lower() in employee.name.lower():
            found = True
            showinfo("Search result", f"{employee.name} - Department: {employee.department} - Birthday: {employee.birthday} - Address: {employee.address} - Onboard: {employee.onboard}")
            break  
    if not found:
        showinfo("Result", "Employee not found!")

#update employee
def update_employee():
    department_value = department.get()
    name_value = name.get()
    birthday_value = birthday.get()
    address_value = address.get()
    onboard_value = onboard.get()
  
    for emp in employee_list:
        if name_value == emp.name:
            emp.department = department_value
            emp.birthday = birthday_value
            emp.address = address_value
            emp.onboard = onboard_value
            
            showinfo("Result", "update employee sucessfull!")

#delete thông tin
def delete_employee():
    search_value = search.get().strip()  # 
    for i, emp in enumerate(employee_list):
        if search_value.lower() == emp.name.lower().strip():  
            del employee_list[i]
            showinfo("Result", "Delete employee successful!")
            return
    showinfo("Result", "Employee not found!")
    
#tạo các button
add_btn = Button(window, text = "Add", command = add_employee)
add_btn.grid(row=7, column=0, columnspan=10)

update_btn = Button(window, text = "Update", command = update_employee)
update_btn.grid(row=7, column=1, columnspan=10)

delete_btn = Button(window, text = "Delete", command = delete_employee)
delete_btn.grid(row=7, column=2, columnspan=10)

search_btn = Button(window, text = "Search", command = search_employee)
search_btn.grid(row=6, column=0, columnspan=1)

# Tạo các trường nhập liệu
department = Entry(window, width=25, bd=5, font=('Arial', 15))
department.grid(row=1, column=1, columnspan=3, padx=5, pady=5)
name = Entry(window, width=25, bd=5, font=('Arial', 15))
name.grid(row=2, column=1, columnspan=3, padx=5, pady=5)
birthday = Entry(window, width=25, bd=5, font=('Arial', 15))
birthday.grid(row=3, column=1, columnspan=3, padx=5, pady=5)
address = Entry(window, width=25, bd=5, font=('Arial', 15))
address.grid(row=4, column=1, columnspan=3, padx=5, pady=5)
onboard = Entry(window, width=25, bd=5, font=('Arial', 15))
onboard.grid(row=5, column=1, columnspan=3, padx=5, pady=5)
search= Entry(window, width=25, bd=5, font=('Arial', 15))
search.grid(row=6, column=1, columnspan=3, padx=5, pady=5)

# Chạy vòng lặp chính của ứng dụng
window.mainloop()
```
